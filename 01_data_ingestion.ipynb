## Setup

```

import pandas as pd
import numpy as np
from pathlib import Path

BASE = Path('..') 
RAW  = BASE / 'data' / 'raw'
```

## Load all 10 raw CSVs

```

files = {
    'fund_master':        '01_fund_master.csv',
    'nav_history':        '02_nav_history.csv',
    'aum_by_fund_house':  '03_aum_by_fund_house.csv',
    'monthly_sip':        '04_monthly_sip_inflows.csv',
    'category_inflows':   '05_category_inflows.csv',
    'folio_count':        '06_industry_folio_count.csv',
    'scheme_performance': '07_scheme_performance.csv',
    'investor_txns':      '08_investor_transactions.csv',
    'portfolio_holdings': '09_portfolio_holdings.csv',
    'benchmark_indices':  '10_benchmark_indices.csv',
}

dfs = {}
for key, fname in files.items():
    df = pd.read_csv(RAW / fname, low_memory=False)
    dfs[key] = df
    print(f'{fname}')
    print(f'  Shape : {df.shape}')
    print(f'  Dtypes: {dict(df.dtypes)}')
    print(f'  Head  :\n{df.head(2)}\n')
```

## Explore fund_master

```

fm = dfs['fund_master']
print('Unique Fund Houses:', fm['fund_house'].nunique())
print(fm['fund_house'].value_counts())
print()
print('Categories:', fm['category'].unique())
print('Risk Grades:', fm['risk_category'].unique())
print('Plans:', fm['plan'].unique())
```

## Validate AMFI code coverage

```

nav_codes  = set(dfs['nav_history']['amfi_code'].unique())
fund_codes = set(dfs['fund_master']['amfi_code'].unique())
print(f'fund_master codes : {len(fund_codes)}')
print(f'nav_history codes : {len(nav_codes)}')
print(f'Missing in nav    : {fund_codes - nav_codes}')
print(f'Extra in nav      : {nav_codes - fund_codes}')
```

## Data quality summary

```

print(f'{'Dataset':<25} {'Rows':>8} {'Cols':>6} {'Nulls':>8}')
print('-'*50)
for key, df in dfs.items():
    print(f'{key:<25} {len(df):>8,} {df.shape[1]:>6} {df.isnull().sum().sum():>8}')
```
